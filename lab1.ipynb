```python
'''Начав тренировки, спортсмен в первый день пробежал 10 км. Каждый следующий день он увеличивал дневную норму на 10% от нормы предыдущего дня. Определить:
а) какой суммарный путь пробежит спортсмен за 7 дней;
б) через сколько дней спортсмен пробежит суммарный путь 100 км;
в) через сколько дней спортсмен будет пробегать в день больше 20 км?'''
days=1
a=10
r=0.1
s=10
#Пункт а)
while days<7:
    days+=1
    a=a+a*r
    s+=a
print('Ответ А:',round(s,2), 'км')
#Пункт б)
days=1
s=10
a=10
r=0.1
while s<100:
    days+=1
    a=a*(1+r)
    s+=a
print('Ответ Б:',days,'дней')
#Пункт в)
days=1
s=10
a=10
r=0.1
while a<20:
    days+=1
    a=a*(1+r)
print('Ответ В:',days,'дней')
    
```

```python
import math
a = 0.1
b = 1
h = 0.1
e = 0.0001
m = []
while a < b:
    m.append(round(a,1))
    a += h
for x in range(len(m)):
    s = 0
    y = (1 + 2*m[x]**2) * math.exp(m[x]**2)
    for i in range(100):
        s1 = (2*i + 1)*m[x]**(2*i) / math.factorial(i)
        if abs(s1) > e:
            s += s1
    print(s,y)
```
